```python
import urllib2
import re
import lxml.html
```

```python
story_url = 'https://www.literotica.com/s/a-brothers-memories-part-4'
```

```python
def download(url, user_agent='liter', num_retries=2):
    print('Downloading', url)
    headers = {'User-agent': user_agent}
    request = urllib2.Request(url, headers=headers)
    try:
        html = urllib2.urlopen(request).read()
    except urllib2.URLError as e:
        print 'Download error:', e.reason
        html = None
        if num_retries > 0:
            if hasattr(e, 'code') and 500 <= e.code < 600:
                #recursively retry 5xx HTTP errors
                return download(url, user_agent, num_retries-1)   
    return html
```

```python
html = download(story_url)
```

```python
tree = lxml.html.fromstring(html)
pages = int(re.match('(\d+)',tree.xpath('//div[@class="b-pager-pages"]/span[@class="b-pager-caption-t r-d45"]/text()')[0]).group(0))
```

```python
story_text = []
for i in range(pages):
    page_url = story_url + '?page=%s' % (i+1)
    page_html = download(page_url)
    page_tree = lxml.html.fromstring(page_html)
    page_text = page_tree.xpath('//div[@class="b-story-body-x x-r15"]/div/p/text()')
    story_text += page_text
    
print story_text
```

```python
from __future__ import division
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
import string
import nltk, re, pprint
from nltk import word_tokenize
```

```python
stopwords = nltk.corpus.stopwords.words('english')
stopwords.extend(["n't","'s","''","``"])
punc = string.punctuation
lemma = WordNetLemmatizer()
p_stemmer = PorterStemmer()

tokens = word_tokenize(" ".join([sent.lower() for sent in story_text]))

content = [word for word in tokens if word not in stopwords and word not in punc]
lemmad = [lemma.lemmatize(word) for word in content]
stemmed = [p_stemmer.stem(word) for word in content]
```

```python
print(len(lemmad))
lemmad1 = lemmad[:1000]
lemmad2 = lemmad[1000:2000]
lemmad3 = lemmad[2000:3000]
lemmad4 = lemmad[3000:]
texts = []
texts.append(lemmad1)
texts.append(lemmad2)
texts.append(lemmad3)
texts.append(lemmad4)
```

```python
import gensim
from gensim import corpora, models
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = 30, id2word = dictionary, passes = 10)
```

```python
topics = ldamodel.print_topics(num_topics=30, num_words=5)
for i in topics:
    print i
```
